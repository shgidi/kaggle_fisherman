```python
#make training annotation files for SSD
#f1=open('./ssd_notati.txt', 'w+')
notations={}
l1=pd.DataFrame()
for fl in lab_files:
    labels=pd.read_json(model_path+'/annos/'+fl)
    
    for i in range(len(labels)):
        img_filename=labels.iloc[i,2]
        print (img_filename)
        print(img_filename.split('/')[-1])
        l1 = pd.DataFrame((labels[labels.filename==img_filename].annotations).iloc[0])
        im=Image.open('/mnt/kaggle_fisherman/train/'+fl.split('_')[0].upper()+'/'+img_filename.split('/')[-1])
        size=im.size
        for row in l1.iterrows():
            x_min=int(row[1][3])
            y_min=int(row[1][4])
            x_max=int(row[1][2])+x_min
            y_max=int(row[1][1])+y_min
            notations[fl.split('_')[0].upper()+'/'+img_filename]=[x_min/size[0],y_min/size[1],x_max/size[0],y_max/size[1],1]
```
